```python
# Creat A Pie Chart of Totals By Country
import matplotlib.pyplot as plt

# Create a new column 
df['order_amount'] = df['Quantity'] * df['Price']

"""
The groupby operation in pandas is a core function used for data analysis, enabling you to split data into groups based on some criteria, apply a function to each group independently, and combine the results into a new data structure. This process is known as the "split-apply-combine" strategy and is similar to the GROUP BY clause in SQL.
"""
country_totals = df.groupby('Country')['order_amount'].sum()
plt.figure(figsize=(8,8))
# Use %% to display a literal '%' character
plt.pie(country_totals, labels=country_totals.index, autopct='%1.1f%%')

plt.title('Total Order Amount by Country')
plt.axis('equal')  # Ensures the pie chart is a circle
plt.show()
```
